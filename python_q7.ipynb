```python
from urllib.request import urlopen
import json
import pandas as pd
url = 'https://data.nasa.gov/resource/y77d-th95.json'
response = urlopen(url)
data_json = json.loads(response.read())
df = pd.json_normalize(data_json)

change_type = {'mass' : float, 
               'reclat' :float,
               'reclong' : float}
df = df.astype(change_type)
df.drop([':@computed_region_cbhk_fwbd', ':@computed_region_nnqa_25f4'], axis = 1 , inplace = True)
```

```python
from datetime import datetime
import numpy as np

df['year'].fillna('1800-01-01T00:00:00.000' ,inplace = True )
for id , d in enumerate(df['year']):    
    if d is not None or d is not Nan:
         df.loc[id , 'year']= datetime.strptime(str(d) , '%Y-%m-%dT%H:%M:%S.%f')
# df['year'] = pd.to_datetime(df['year'] , format='%Y-%m-%d %H:%M:%S')
# We cant convert column do  datetime64(ns) , due to out of range values on row 174 , which is causing error. Pandas do not support outof bond values

```
