### Q-2. Imagine you have a dataset where you have different features like Age,Gender , Height , Weight , BMI , and Blood Pressure and you have to classify the people into different classes like Normal , Overweight , Obesity , Underweight , and Extreme Obesity by using  any 4 different classification algorithms. Now you have to build a model which can classify people into different classes.

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('dataset/Obesity.csv')
df.head(10)
```

```python
df.info()
```

We dont any any null values

```python
for i in df.columns:
    print(len(df[i].unique()) , i , df[i].dtype)
    
```

```python
df.columns
```

```python
y = df['NObeyesdad']
if 'NObeyesdad' in df.columns:
    df.drop('NObeyesdad' , inplace = True , axis = 1)
    
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

#### we will apply onehotencoder on every categorical columns

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output= False)
df_cat = ohe.fit_transform(df.select_dtypes(include = 'O'))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_cat = sc.fit_transform(df_cat)
```

```python
df_cat = pd.DataFrame(df_cat, columns = ohe.get_feature_names_out())
```

```python



df_num = sc.fit_transform(df.select_dtypes(include = 'number'))
```

```python
df_num = pd.DataFrame(df_num , columns = sc.get_feature_names_out())
df_num
```

```python
df_transformed = pd.concat([df_num , df_cat] , axis = 1)
df_transformed
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X_train , X_test , y_train , y_test = train_test_split(df_transformed , y, test_size = 0.2 , random_state = 45)
# Define the parameter grids for each model
random_forest_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
}

logistic_regression_params = {
    'C': [0.1, 1.0, 10.0],
    'solver': ['liblinear', 'saga'],
    'max_iter':[1000000]
}

svm_params = {
    'C': [1.0, 10.0, 100.0],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto'],
    'max_iter':[1000000]
}

decision_tree_params = {
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
}

knn_params = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
}

# Generate random parameters for each model


# Create GridSearchCV objects for each model
random_forest_grid_search = GridSearchCV(RandomForestClassifier(), random_forest_params , cv = 5 , scoring = 'accuracy')
logistic_regression_grid_search = GridSearchCV(LogisticRegression(), logistic_regression_params , cv = 5,  scoring = 'accuracy')
svm_grid_search = GridSearchCV(SVC(), svm_params , cv = 5,   scoring = 'accuracy')
decision_tree_grid_search = GridSearchCV(DecisionTreeClassifier(), decision_tree_params , cv =5,  scoring = 'accuracy')
knn_grid_search = GridSearchCV(KNeighborsClassifier(), knn_params ,cv = 5,  scoring = 'accuracy')



# Print the random parameter grids for each model
print("Random Parameter Grid for Random Forest:")
print(random_forest_grid_search.param_grid)
print()

print("Random Parameter Grid for Logistic Regression:")
print(logistic_regression_grid_search.param_grid)
print()

print("Random Parameter Grid for SVM:")
print(svm_grid_search.param_grid)
print()

print("Random Parameter Grid for Decision Tree:")
print(decision_tree_grid_search.param_grid)
print()

print("Random Parameter Grid for KNN:")
print(knn_grid_search.param_grid)
```

```python
# Lets start Training
models = [random_forest_grid_search, random_forest_grid_search, logistic_regression_grid_search ,svm_grid_search, decision_tree_grid_search, knn_grid_search]
for model in models:
    model.fit(X_train , y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test , y_pred)
    print(f'{model.best_estimator_ ,model.best_params_ } \n')
    print(f'Training Score : {model.best_score_ } , Validation Score : {score} \n')
```

## As per results we can creat model of svm , and deploy 

```python
svc = SVC(C= 100.0, gamma= 'scale', kernel= 'linear', max_iter= 1000000)
svc.fit(X_train , y_train)
y_pred = svc.predict(X_test)
score = accuracy_score(y_pred , y_test)
```
